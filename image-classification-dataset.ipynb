# Classification dataset #

```python
import os
import shutil
```

```python
# Make a directory for the dowloaded datasets
my_dataset_path = '../data/my-data/my-classification-dataset'
os.mkdir(my_dataset_path)
```

```python
# Subfolders for each emotional class within dataset

os.mkdir( os.path.join(my_dataset_path, 'neutral-face') )
os.mkdir( os.path.join(my_dataset_path, 'happy-face') )
os.mkdir( os.path.join(my_dataset_path, 'sad-face') )
os.mkdir( os.path.join(my_dataset_path, 'angry-face') )
os.mkdir( os.path.join(my_dataset_path, 'scared-face') )
os.mkdir( os.path.join(my_dataset_path, 'surprised-face') )
os.mkdir( os.path.join(my_dataset_path, 'disgusted-face') )
```

Navigate to pinterest and dowload relevant boards for each emotional state:

```python
# Neutral face (>100 images)

!gallery-dl https://www.pinterest.co.uk/bodiebodes/portraits/female-front/, https://www.pinterest.co.uk/asiavinae/neutral-front-faces/, https://www.pinterest.co.uk/f1nd1ngp4nd0ra/neutral-faces/, https://www.pinterest.co.uk/bodiebodes/portraits/male-front/
```

```python
# Add to respective emotional class folder (>500)

source_directory = '../My-work/gallery-dl' 
class_folder = 'neutral-face'
target_directory = os.path.join(my_dataset_path, class_folder)
    
file_names = os.listdir(source_directory)
    
for file_name in file_names:
    shutil.move(os.path.join(source_directory, file_name), target_directory)
```
